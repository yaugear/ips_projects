```python
def get_disamed_indexes(contexts):
    result = []
    chains = []
    for context in contexts:
        morph = context[2][1]
        if morph not in chains:
            chains.append(morph)
        result.append(chains.index(morph))
    return result, chains
```

```python
def printc(context):
    result = u''
    for row in context:
        if len(row) == 2:
            result += " %s(%s)" % row
    return result
```

```python
print(printc(sentences1[0]))
```

```python
def save_as(filename, contexts, labels):
    cons = []
    cons = [printc(context) for context in contexts]
    result = '\n'.join(['%s\t%s' % row for row in zip(cons, labels)])
    
    with open(filename, 'wb') as stream:
        stream.write(result.encode('utf-8'))
        
    print('Done!')
```

## KMeans Clustering

#### KMeans with 2 Cluster and without encoder

```python
def get_vector_long(context):
    return chain2bin(context[0][1] if context[0][1] else '') + \
           chain2bin(context[1][1] if context[1][1] else '') + \
           chain2bin(context[3][1] if context[3][1] else '') + \
           chain2bin(context[4][1] if context[4][1] else '')
```

```python
def get_vector_short(context):
    return get_smaller_repr(chain2bin(context[0][1] if context[0][1] else '')) + \
           get_smaller_repr(chain2bin(context[1][1] if context[1][1] else '')) + \
           get_smaller_repr(chain2bin(context[3][1] if context[3][1] else '')) + \
           get_smaller_repr(chain2bin(context[4][1] if context[4][1] else ''))
```

```python
def collect_stat_as_table(labels, true_indexes, chains):
    comp = zip(labels, true_indexes)
    clusters = {}
    
    all_sample_count = len(labels)

    for (cluster, chain_index) in comp:
        chain = chains[chain_index]
        if cluster not in clusters:
            clusters[cluster] = {chain: 1}
        else:
            if chain not in clusters[cluster]:
                clusters[cluster][chain] = 1
            else:
                clusters[cluster][chain] += 1
    chs = []
    for value in set(true_indexes):
        chs.append((chains[value], true_indexes.count(value)))
    
    avgd = 0
    for cluster in clusters:
        sample_count = sum([clusters[cluster][chain] for chain in chains if chain in clusters[cluster]])
        dominant = sorted([(chain, clusters[cluster][chain]) for chain in chains if chain in clusters[cluster]], key=lambda x: -x[1])[0]
        clusters[cluster]['3. Most Freq Chain'] = dominant[0]
        clusters[cluster]['1. Samples'] = sample_count
        clusters[cluster]['4. % of Dominant Chain in the Cluster'] = 100. * dominant[1] / sample_count
        clusters[cluster]['2. % of All Samples'] = 100. * sample_count / all_sample_count
        avgd += 100. * dominant[1] / all_sample_count
    
#     avgd = avgd / len(clusters)
    clusters['Sum'] = dict(chs)
    clusters['Sum']['2. % of All Samples'] = 100
    clusters['Sum']['1. Samples'] = all_sample_count
    clusters['Sum']['4. % of Dominant Chain in the Cluster'] = 'Avg: %%%.2f' % avgd
    
    return pd.DataFrame(clusters).transpose()
```

```python
from sklearn.cluster import KMeans
```

```python
model2 = KMeans(n_clusters=2)
```

```python
X = [get_vector_long(context) for context in sentences1]
```

```python
model2.fit(X)
```

```python
collect_stat_as_table(model2.labels_, true_indexes1, chains1)
```

```python
save_as(u'әле_2_table1.csv', sentences1, model2.labels_)
```
